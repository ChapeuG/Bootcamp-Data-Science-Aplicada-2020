####Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado.

```
# separar o indicador de reião presente no nome de cada estado
usaveis['Região'] = usaveis.index.str[0]

# criar um dicionário com as regiões do pais
regioes = {1:'Norte', 2:'Nordeste', 3:'Sudeste', 4:'Sul', 5:'Centro Oeste'}

# aplicar o código criado no dicionário para cada estado
usaveis['Região'] = usaveis['Região'].map(lambda i: regioes[int(i)])
usaveis['Região'].head(2)
```

map() é uma função builtin de Python, isto é, uma função que é implementada diretamente no interpretador Python, podendo ser utilizada sem a importação de um módulo específico. Essa função, em Python, serve para aplicarmos uma função a cada elemento de uma lista, retornando uma nova lista contendo os elementos resultantes da aplicação da função. Considere o exemplo abaixo:
